```python
# Function to calculate continuous feature statistics

def continuous_features_report(df):
  continuous_report = {}
  for col in df.select_dtypes(include=['number']).columns:
    col_data = df[col]
    continuous_report[col] = {
      'count': col_data.count(),
      '% missing': col_data.isna().mean()*100,
      'cardinality': col_data.nunique(),
      'min': col_data.min(),
      '1st quartile': col_data.quantile(0.25),
      'mean': col_data.mean(),
      'median': col_data.median(),
      '3rd quartile': col_data.quantile(0.75),
      'max': col_data.max(),
      'std.dev. ': col_data.std()
    }
  return pd.DataFrame(continuous_report).T

# Function to calculate categorical feature statistics
def categorical_features_report(df):
  categorical_report = {}
  for col in df.select_dtypes(include=['object', 'category']).columns:
    col_data = df[col]
    mode_value = col_data.mode()[0] if not col_data.mode().empty else None
    mode_freq = col_data.value_counts().iloc[0] if mode_value else 0
    second_mode = col_data.value_counts().index[1] if len(col_data.value_counts()) > 1 else None
    second_mode_freq = col_data.value_counts().iloc[1] if second_mode else 0

    categorical_report[col] = {
      'count': col_data.count(),
      '% missing': col_data.isna().mean()*100,
      'cardinality': col_data.nunique(),
      'mode': mode_value,
      'mode freq': mode_freq,
      'mode %': (mode_freq/col_data.count())*100 if col_data.count() else 0,
      '2nd mode': second_mode,
      '2nd mode freq': second_mode_freq,
      '2nd mode %': (second_mode_freq/col_data.count())*100 if col_data.count() else 0
    }
  return pd.DataFrame(categorical_report).T

# Generate report

continuous_report = continuous_features_report(data1)
categorical_report = categorical_features_report(data1)

# Display the reports

print("Continuous Feature Report:")
print(continuous_report)

print("\nCategorical Feature Report:")
print(categorical_report)
```

Data Quality

```python
import pandas as pd

# Function to identify missing values
def missing_values_report(df):
    missing_report = pd.DataFrame(df.isnull().sum(), columns=['Missing Count'])
    missing_report['% Missing'] = (df.isnull().mean() * 100).round(2)
    return missing_report[missing_report['Missing Count'] > 0]

# Function to identify irregular cardinality
def cardinality_report(df, threshold=10):
    cardinality_report = {}
    for col in df.columns:
        unique_values = df[col].nunique()
        if unique_values < threshold:
            cardinality_report[col] = unique_values
    return pd.DataFrame(list(cardinality_report.items()), columns=['Feature', 'Unique Values'])

# Function to identify outliers using IQR method for continuous columns
def outliers_report(df, threshold=1.5):
    outliers = {}
    for col in df.select_dtypes(include=['number']).columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - threshold * iqr
        upper_bound = q3 + threshold * iqr
        outliers_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
        outliers[col] = outliers_count
    return pd.DataFrame(list(outliers.items()), columns=['Feature', 'Outlier Count'])

# Function to create the Data Quality Plan (like Table 3.5)
def data_quality_plan(df):
    # Summary of missing values
    missing_report = missing_values_report(df)

    # Summary of cardinality (columns with less than threshold unique values, e.g., 10)
    cardinality_report_df = cardinality_report(df)

    # Summary of outliers
    outliers_report_df = outliers_report(df)

    # Combine all summaries
    quality_issues = []
    for col in df.columns:
        issue = {
            'Feature': col,
            'Missing Count': missing_report.loc[col, 'Missing Count'] if col in missing_report.index else 0,
            '% Missing': missing_report.loc[col, '% Missing'] if col in missing_report.index else 0,
            'Cardinality': cardinality_report_df.set_index('Feature').get('Unique Values', {}).get(col, df[col].nunique()),
            'Outlier Count': outliers_report_df.set_index('Feature').get('Outlier Count', {}).get(col, 0)
        }
        quality_issues.append(issue)

    # Create a DataFrame for the Data Quality Plan
    data_quality_df = pd.DataFrame(quality_issues)
    data_quality_df['Cardinality Issue'] = data_quality_df['Cardinality'] < 10  # Mark irregular cardinality
    return data_quality_df

# Run the data quality checks
data_quality_df = data_quality_plan(data1)

# Display the Data Quality Plan
print("Data Quality Plan:")
print(data_quality_df)
```
