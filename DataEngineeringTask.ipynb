```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL for scraping
url = "https://www.booking.com/searchresults.en-gb.html?aid=397594&label=gog235jc-1DCAEoggI46AdICVgDaFCIAQGYAQm4ARfIAQzYAQPoAQH4AQKIAgGoAgO4ArDuuaEGwAIB0gIkZmJhYjE4YzAtNDdhMy00MmY1LTk2NWItN2UzOTgyNTk1OWEx2AIE4AIB&sid=0e1245fc48516aae83900b2f39a1c320&checkin=2025-03-01&checkout=2025-03-02&region=1000&"
```

```python
# Step 1: Scrape the data
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Extracting "Price" and "Room Type"
prices = []
room_types = []

# Locate elements using classes or tags
# (Update these selectors based on the structure found in your inspection)

price_elements = soup.find_all("div", class_="f6431b446c fbfd7c1165 e84eb96b1f")


for div in price_elements:
    price_span = div.find("span", class_="f6431b446c fbfd7c1165 e84eb96b1f")
    if price_span:
        price = price_span.get_text(strip=True)
        print(price)

room_elements = soup.find_all("h4", class_="abf093bdfe e8f7c070a7")  # Finding all h4 elements with the class for room types

for room in room_elements:
    room_types.append(room.get_text(strip=True))  # Extract text and strip extra spaces
```

```python
# Step 2: Save data into a DataFrame
data = {
    "Price": prices,
    "Room Type": room_types
}
df = pd.DataFrame(data)

df
```

```python
# Step 3: Perform data cleaning/transformation
# - Remove empty or null entries
df.dropna(inplace=True)
# - Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Step 4: Save to CSV
csv_file_path = "Cleaned_Data.csv"
df.to_csv(csv_file_path, index=False)
print(f"Data saved to {csv_file_path}")
```
