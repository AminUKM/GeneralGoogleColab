```python
import scrapy

class HotelSpider(scrapy.Spider):
    name = 'hotel_spider'
    start_urls = [
        "https://www.booking.com/searchresults.en-gb.html?aid=397594&label=gog235jc-1DCAEoggI46AdICVgDaFCIAQGYAQm4ARfIAQzYAQPoAQH4AQKIAgGoAgO4ArDuuaEGwAIB0gIkZmJhYjE4YzAtNDdhMy00MmY1LTk2NWItN2UzOTgyNTk1OWEx2AIE4AIB&sid=0e1245fc48516aae83900b2f39a1c320&checkin=2025-03-01&checkout=2025-03-02&region=1000&"
    ]

    def parse(self, response):
        # Extracting Prices
        prices = response.css("div.e84eb96b1f.a661120d62 span[data-testid='price-and-discounted-price']::text").getall()

        # Extracting Room Types
        room_types = response.css("h4.abf093bdfe.e8f7c070a7::text").getall()

        # Yielding the extracted data
        for price, room_type in zip(prices, room_types):
            yield {
                'Price': price.strip(),
                'Room Type': room_type.strip()
            }
```
